```python
import pandas as pd
from src.visualisation import visualize_spectrogram
from src.preprocessing import transform_wav_to_spectrogram, preprocessing_spectrograms
import torchaudio.transforms as T
```

```python
paths = ['data/train/' + n for n in ['00000.wav', '00050.wav', '00100.wav', '01000.wav', '02000.wav', '10000.wav']]
spectrogram = T.Spectrogram(n_fft=512, hop_length=128)
amplitude_to_db = T.AmplitudeToDB(stype='power', top_db=80)
specs = [(transform_wav_to_spectrogram(path, spectrogram, amplitude_to_db), path.split('/')[-1]) for path in paths]
specs
```

```python
for spec, name in specs:
    visualize_spectrogram(spec, name)
```
